```python
import math
import random
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D 
```

## Random Walk in 1D

```python
ite=[100,200,300]
```

```python
s=0
```

```python
displacements = []
for n in ite:
    walk = []
    displacement=[]
    for i in range(n):
        if (random.random())>=0.5:
            walk.append(1)
        else:
            walk.append(-1)
        displacement.append(sum(walk))
    displacements.append(displacement)
```

```python
plt.plot(range(ite[0]),displacements[0],color='red')
plt.plot(range(ite[1]),displacements[1],color='blue')
plt.plot(range(ite[2]),displacements[2],color='green')
```

## Random Walk in 2D

```python
n=1000
x = np.zeros(n) 
y = np.zeros(n) 
directions=["NORTH","SOUTH","EAST","WEST"] # Assuming the four directions of movement.
for i in range(1,n):
    step_direction = random.choice(directions)
    if step_direction == "EAST": #updating the direction with respect to the direction of motion choosen.
        x[i] = x[i - 1] + 1
        y[i] = y[i - 1] 
    elif step_direction == "WEST": 
        x[i] = x[i - 1] - 1
        y[i] = y[i - 1] 
    elif step_direction == "NORTH": 
        x[i] = x[i - 1] 
        y[i] = y[i - 1] + 1
    else: 
        x[i] = x[i - 1] 
        y[i] = y[i - 1] - 1
```

```python
dispalcement = [math.sqrt(a**2)+(b**2) for a,b in zip(x,y)]
```

```python
plt.plot(range(n),dispalcement)
plt.title('Dispalcement Plot')
plt.show()
```

```python
plt.plot(x,y)
plt.title('Random Walk in 2D')
plt.show()
```

## Random Walk in 3D

```python
N =1500
R = (np.random.rand(N)*6).astype("int") 
x = np.zeros(N) 
y = np.zeros(N)
z = np.zeros(N)
x[ R==0 ] = -1; x[ R==1 ] = 1 
y[ R==2 ] = -1; y[ R==3 ] = 1
z[ R==4 ] = -1; z[ R==5 ] = 1
x = np.cumsum(x) 
y = np.cumsum(y)
z = np.cumsum(z)
plt.figure()
ax = plt.subplot(1,1,1, projection='3d')
ax.plot(x, y, z,alpha=0.9) #alpha sets the darkness of the path.
ax.scatter(x[-1],y[-1],z[-1])
plt.show()
```

```python
dispalcement = [math.sqrt((a**2)+(b**2)+(c**2)) for a,b,c in zip(x,y,z)]
```

```python
plt.plot(range(N),dispalcement)
plt.title('Dispalcement Plot')
plt.show()
```

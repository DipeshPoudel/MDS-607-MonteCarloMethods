```python
import random
```

```python
n=10000
```

```python
s=0
```

```python
walk = []
displacement=[]
for i in range(n):
    if (random.random())>=0.5:
        walk.append(1)
    else:
        walk.append(-1)
    displacement.append(sum(walk))
```

```python
import matplotlib.pyplot as plt
plt.plot(range(n),displacement)
plt.figure(figsize=((25), (15)))
plt.show()
```

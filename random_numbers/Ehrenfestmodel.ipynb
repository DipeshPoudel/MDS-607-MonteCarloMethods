# Simulation of Ehrenfest model

Do a computer simulation of the Markov chain for N = 100. Start from state 0 (one of the partitions is empty) and follow the chain up to 1000 steps. Draw a graph of the number of molecules in the initially empty partition as a function of the number of steps. On the basis of your simulation, would you expect to observe during the course of the simulation a return to state 0?

```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
```

```python
N=100
urn_1_steps = []
urn_2_steps = []
```

```python
n_rept = 1000
```

```python
box = [[0,1] for a,b in zip((range(N)),(range(N)))]
```

```python
for i in range(n_rept):
    x = random.randint(N)
    box[x][0],box[x][1]=box[x][1],box[x][0]
    box_a = np.array(box,dtype=int)
    total = np.sum(box_a,axis=0)
    urn_1_steps.append(total[0])
    urn_2_steps.append(total[1])
```

```python
plt.plot(range(n_rept),urn_1_steps,label="Urn 1")
plt.plot(range(n_rept),urn_2_steps, label="Urn 2")
plt.title('Number of Molecules in Each Urn in Each Step')
plt.xlabel('Number of Steps')
plt.ylabel('Number of Molecules')
plt.legend()
```

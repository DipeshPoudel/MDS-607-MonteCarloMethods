```python
N=10000
errors = []
random_number_list = [random.random() for i in range(N)]
function_value = []
for num,xi in enumerate(random_number_list):
    function_value.append(math.exp(xi))
    mean_value = sum(function_value)/(num+1)
    errors.append(abs(np.array(mean_value)-true_value))
```

```python
plt.plot(errors)
```

```python
N=10000
errors = []
std_errors = []
random_number_list = [random.random() for i in range(N)]
function_value = []
for num,xi in enumerate(random_number_list):
    function_value.append(math.exp(xi))
    mean_value = sum(function_value)/(num+1)
    errors.append(abs(np.array(mean_value)-true_value))
    std_errors.append((np.std(function_value))/math.sqrt(num+1))
```

```python
plt.plot(errors)
plt.plot(std_errors)
```

## Estimation of Pie

```python
import random
```

```python
circle_points = 0
square_points = 0
for i in range(1000):
    x = random.random()
    y = random.random()
    d = ((x*x)+(y*y))
    if d<=1:
        circle_points+=1
    square_points+=1
    pie = 4*(circle_points/square_points)
    print(pie)
```

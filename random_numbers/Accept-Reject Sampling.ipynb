```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
import math
np.random.seed(523)
```

```python
n=10000
```

```python
def f1(x):
    return np.sin(x-0.5)**2 * np.exp(-x**2) + .2*np.exp(-x**2)
```

```python
x = np.linspace(-3,3,n)
y = f1(x)
plt.plot(x,y)
plt.title("The Function")
```

```python
def accept_reject(N):
    xmin = -3
    xmax = 3
    pmax = 0.8

    n_accept=0
    x_list = [] 
    while n_accept < N:
        t = (xmax-xmin)*np.random.rand() + xmin
        y = np.random.rand()
        if y < f1(t)/ pmax:
            n_accept += 1
            x_list.append(t)
    return x_list
```

```python
x = accept_reject(10000)
bins, edges, patches = plt.hist(x, bins=100)
```
